```python
import pandas as pd
import matplotlib.pyplot as plt

#Graph Parameters
plt.rcParams['figure.figsize'] = 12,8
plt.rc('text', usetex=False)
plt.rcParams.update({'font.size': 15,
                     'legend.fontsize': 15})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['lines.linewidth'] = 1.5
```

```python
# Running the python script to generate csv data (using ipython magic)
# This would also produce below any printing output

%run kepler.py
```

```python
# Loading CSV Files and converting into Pandas dataframe

rRK, vRK, t = pd.read_csv('RKData/rRK.csv'), pd.read_csv('RKData/vRK.csv'), pd.read_csv('RKData/t.csv')
rEC, vEC, t = pd.read_csv('EulerData/rEC.csv'), pd.read_csv('EulerData/vEC.csv'), pd.read_csv('EulerData/t.csv')
```

```python
# Unpacking into separate dimensions

xRK, yRK, zRK = rRK.iloc[:, 0], rRK.iloc[:, 1], rRK.iloc[:, 2]
xEC, yEC, zEC = rEC.iloc[:, 0], rEC.iloc[:, 1], rEC.iloc[:, 2]
```

```python
# Initializing Figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 10))
plt.suptitle("Simulation Duration = 100 Orbital Periods")

# Plotting RK4 Trajectory
ax1.plot(xRK, yRK, label="RK Trajectory")
ax1.set_xlabel('X Position [AU]')
ax1.set_ylabel('Y Position [AU]')
ax1.set(aspect=1)
ax1.set_title("Trajectory Integrated by RK4")

# Plotting Euler Trajectory
ax2.plot(xEC, yEC, label="Euler Trajectory")
ax2.set_xlabel('X Position [AU]')
ax2.set_ylabel('Y Position [AU]')
ax2.set(aspect=1)
ax2.set_title("Trajectory Integrated by Euler-Cromer")
```
